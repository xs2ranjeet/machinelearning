```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import dataset
flatdata = pd.read_csv('HomeBuyer.csv')
X = flatdata.iloc[:,[0,1]].values
y = flatdata.iloc[:,2].values
```

```python
#split dataset into train and test(75:25) ratio
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size= 1/4, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
independent_scaler = StandardScaler()
X_train = independent_scaler.fit_transform(X_train) # fit and trasnform
X_test = independent_scaler.transform(X_test) #transform only
```

```python
#Find logistic regression for training data.Logistic regression is a 
# linear classifier after which the data is divided by a straight line.
from sklearn.linear_model import LogisticRegression
LRClassifier = LogisticRegression(random_state = 0)
LRClassifier.fit(X_train, y_train)
```

```python
#predict the values
prediction = LRClassifier.predict(X_test)
```

```python
#Create confusion matrix to evalute performance of data
from sklearn.metrics import confusion_matrix
confusionMatrix  = confusion_matrix(y_test, prediction)
```

```python
print(confusionMatrix)
```

```python
## let us visualize it
```

```python
xx, yy = np.mgrid[-5:5:.01, -5:5:.01]
grid = np.c_[xx.ravel(), yy.ravel()]
probs = LRClassifier.predict_proba(grid)[:,1].reshape(xx.shape)
print(probs)
```

```python
f, ax = plt.subplots(figsize = (8,6))
contour = ax.contourf(xx, yy, probs, 25, cmap = "RdBu", vmin = 0, vmax=1)
ax_c = f.colorbar(contour)
ax_c.set_label("$P(y = 1)$")
ax_c.set_ticks([0,.25,.5,.75,1])

ax.scatter(X_test[:,0], X_test[:,1], c = (y_test == 1), s = 50,
          cmap = "RdBu", vmin = -.2, vmax = 1.2, edgecolor = "white", linewidth =1)

ax.set(aspect="equal", xlim = (-5,5), ylim=(-5,5), xlabel="$X_1$", ylabel="$X_2$")

plt.show()
```
